```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error
```

```python
data = pd.read_csv('Housing.csv')
```

```python
data
```

```python
data.dropna()
```

```python
data['mainroad'] = data['mainroad'].map({'yes':1,'no':0})
```

```python
data
```

```python
data['guestroom'] = data['guestroom'].map({'yes':0.1,'no':0.0})
```

```python
data
```

```python
data['basement'] = data['basement'].map({'no':0,'yes':1})
```

```python
data
```

```python
data['hotwaterheating'] = data['hotwaterheating'].map({'no':0,'yes':1})
```

```python
data
```

```python
data['airconditioning'] = data['airconditioning'].map({'yes':1,'no':0})
data
```

```python
data['furnishingstatus'].value_counts()
```

```python
data['furnishingstatus'] = data['furnishingstatus'].map({'unfurnished':0.0,'semi-furnished':0.1,'furnished':0.2})
data
```

```python
data['prefarea'] = data['prefarea'].map({'yes':1,'no':0})
```

```python
x = data[['area','bedrooms','bathrooms','stories','parking','mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']]
y = data[['price']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_absolute_percentage_error(y_test,y_pred)
```

```python
mse
```
